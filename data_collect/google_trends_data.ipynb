# Google Trends API

Selenium으로 작업이 너무 오래걸리고 불편해서 API 사용

`kw_list`(키워드_목록): 최대 5개 길이의 키워드를 리스트로 제공  
`cat`(카테고리): [카테고리 코드는 Github](https://github.com/pat310/google-trends-api/wiki/Google-Trends-Categories) 확인  
`geo`(지역): `''`(기본값): 전 세계, `US`: 미국, `KR`: 한국, `US-CA`: 미국 캘리포니아  
`timeframe`(기간_범위): `2022-01-01 2022-12-31`, `today 12-m`: 지난 12개월, `now 7-d`: 지난 7일, `now 1-H`: 지난 1시간  
gprop(검색유형): `''`(기본값): 웹 검색, `images`: 이미지, `news`: 뉴스, `youtube`: 유튜브, `froogle`: Google Shopping

| 함수                   | 인자                                                                                 | 설명                                       |
|------------------------|--------------------------------------------------------------------------------------|------------------------------------------|
| `interest_over_time`   | 없음                                                                                | 시간별 관심도 데이터를 반환               |
| `interest_by_region`   | `resolution='COUNTRY': str, inc_low_vol=False: bool, inc_geo_code=False: bool`       | 지역별 관심도 데이터를 반환               |
| `related_queries`      | 없음                                                                                | 관련 검색어 데이터를 반환                 |
| `related_topics`       | 없음                                                                                | 관련 주제 데이터를 반환                   |
| `trending_searches`    | `pn='united_states': str`                                                           | 특정 지역의 트렌드 검색어를 반환          |
| `top_charts`           | `date: int, hl='en-US': str, tz=0: int, geo='': str, cat=0: int`                    | 특정 연도/카테고리의 상위 검색어를 반환    |
| `suggestions`          | `keyword: str`                                                                      | 검색어 자동완성 제안을 반환               |



`interest_over_time()` 시간별 관심도: `kw_list`, `timeframe` 필요  
`interest_by_region()` 지역별 관심도: `kw_list`, `geo` 필요  
`related_queries()` 관련 검색어: `kw_list`  
`related_topics()` 관련 주제: `kw_list`  

`trending_searches()` 전 세계 또는 특정 지역의 트렌드 검색어: `geo`  
`top_charts()` 특정 연도 및 카테고리의 상위 검색어: `geo`, `cat`  
`suggestions()` 자동완성 제안: `kw_list`

```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import koreanize_matplotlib
%matplotlib inline

def diff_keyword_plt(keyword_list, title_text):
    period = "today 5-y"
    google_trends.build_payload(kw_list=keyword_list, timeframe=period, geo="KR")
    trend_df = google_trends.interest_over_time()
    plt.rc('font', family='NanumGothic')
    plt.style.use("ggplot")
    plt.figure(figsize=(14,5))
    for keyword in keyword_list:
        plt.plot(trend_df[keyword], label=keyword)
    plt.title(title_text, size=15)
    plt.legend(loc="best")
    plt.show()
```

```python
import pandas as pd

# date_format = "%-m/%-d/%Y"
# date_format_text = "%m%d%Y"
date_format = "%Y-%m-%d"

# 시작 날짜와 종료 날짜에 대해 월별로 첫날과 마지막 날을 생성
# 각 달의 첫날과 마지막 날을 문자열로 추가
def date_range(start, end):
    date_range = pd.date_range(start=start, end=end, freq='MS')  # 각 달의 시작일
    search_date_list = []
    for date in date_range:
        first_day = date.strftime(date_format)
        last_day = (date + pd.offsets.MonthEnd(0)).strftime(date_format)
        search_date = f"{first_day} {last_day}"
        search_date_list.append(search_date)
    
    return search_date_list

search_date_list = date_range("20190101", "20240501")
print("Search Date List:", search_date_list)
```

# 키워드 단독 검색 (동일 검색어 처리 X)

```python
google_trends = TrendReq(hl='ko', tz=600)
```

```python
keyword_list = ["알러지", "대상포진", "수족구병", "장염", "식중독", "감기", "인플루엔자", "동상"]

interest_df = pd.DataFrame()
interest_by_region_df = pd.DataFrame()

for keyword in keyword_list:
    for search_date in search_date_list:
        google_trends.build_payload(kw_list=[keyword], timeframe=search_date, geo='KR')
        interest_df = pd.concat([interest_df, google_trends.interest_over_time()])

        region_date = search_date[:7]
        temp_df = google_trends.interest_by_region(resolution='CITY', inc_low_vol=True, inc_geo_code=True)
        temp_df["keyword"] = keyword
        temp_df["date"] = region_date
        interest_by_region_df = pd.concat([interest_by_region_df, temp_df])

# 데이터를 피벗하여 각 키워드가 컬럼으로 나뉘게 변환
interest_by_region_df = interest_by_region_df.pivot_table(
    index=["geoCode", "date"],
    columns="keyword",
    aggfunc="first"
).reset_index()

# 컬럼 이름 정리
interest_by_region_df.columns.name = None
# interest_by_region_df = interest_by_region_df.rename(columns={"대상포진": "대상포진", "알러지": "알러지"}, )

interest_df.to_csv("interest_df.csv")
interest_by_region_df.to_csv("interest_by_region_df.csv")
```
