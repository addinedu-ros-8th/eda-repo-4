```python
import pandas as pd
import openpyxl
```

```python
data_2019_01 = pd.read_excel("2019년 01월.xlsx")
data_2019_01.head(30)
```

```python
data_2019_01.info()
```

```python
new_order = ['측정일시', '지역', '망', '측정소코드', '측정소명', '주소', 'SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM25']

data_2019_01 = data_2019_01[new_order]
data_2019_01.head()
```

```python
new_order = ["측정일", "측정소코드", "측정소명", "주소", "지역", "지역_구군", "망", "SO2", "CO", "O3", "NO2", "PM10", "PM25"]

# SO2, CO, O3, NO2, PM10, PM25만 평균값 계산
columns_to_aggregate = ["SO2", "CO", "O3", "NO2", "PM10", "PM25"]
df_aggregated = data_2019_01[columns_to_aggregate].groupby(data_2019_01.index // 24).mean().reset_index(drop=True)

# 나머지 열들은 첫 번째 값을 그대로 가져오기
df_aggregated["지역"] = data_2019_01.groupby(data_2019_01.index // 24)["지역"].first().reset_index(drop=True)
df_aggregated["망"] = data_2019_01.groupby(data_2019_01.index // 24)["망"].first().reset_index(drop=True)
df_aggregated["측정소코드"] = data_2019_01.groupby(data_2019_01.index // 24)["측정소코드"].first().reset_index(drop=True)
df_aggregated["측정소명"] = data_2019_01.groupby(data_2019_01.index // 24)["측정소명"].first().reset_index(drop=True)
df_aggregated["주소"] = data_2019_01.groupby(data_2019_01.index // 24)["주소"].first().reset_index(drop=True)

# "측정일시"는 첫 번째 값만 남기기
df_aggregated["측정일"] = data_2019_01.groupby(data_2019_01.index // 24)["측정일시"].first().astype(str).str[:-2].reset_index(drop=True)

# 구군 단위 지역 컬럼 추가
df_aggregated["지역_구군"] = df_aggregated["지역"]

# 지역 컬럼을 시 단위로 변경
df_aggregated["지역"] = df_aggregated["지역"].apply(lambda x: x.split(" ")[0])

df_aggregated = df_aggregated[new_order]

df_aggregated.head()
```

```python
# 시 단위 데이터
df_aggregated_si = df_aggregated.groupby(['지역', '측정일'])[columns_to_aggregate].mean().reset_index()
df_aggregated_si
```

```python
from glob import glob

air_excel_dir = "/Users/yun/Documents/dev_ws/eda/eda_project/eda_poject_data/air"

excel_files = glob(f"{air_excel_dir}/*.xlsx")
excel_files
```

```python
import pandas as pd

# CSV 파일들을 읽어 리스트로 저장
dfs = [pd.read_excel(excel_file) for excel_file in excel_files]

# 모든 DataFrame을 병합하여 하나의 DataFrame 생성
merged_df = pd.concat(dfs, ignore_index=True)

# 병합된 DataFrame 확인
merged_df.info()
```

```python
rename_info = {"지역": "region", "측정일시": "report_date"}
merged_df.rename(columns=rename_info, inplace=True)
merged_df
```

```python
air_df = merged_df.sort_values(by=['report_date', 'region'], ascending=[True, True]).reset_index(drop=True)
air_df.head()
```

```python
drop_col_list = ['망', '측정소코드', "측정소명", "주소"]
air_df.drop(columns=drop_col_list, inplace=True)
```

```python
air_df["region"] = air_df["region"].apply(lambda x: x.split(" ")[0])
air_df.head()
```

```python
new_order = ["id", "region", "report_date", "SO2", "CO", "O3", "NO2", "PM10", "PM25"]

# SO2, CO, O3, NO2, PM10, PM25만 평균값 계산
columns_to_aggregate = ["SO2", "CO", "O3", "NO2", "PM10", "PM25"]
df_aggregated = air_df[columns_to_aggregate].groupby(air_df.index // 24).mean().reset_index(drop=True)

df_aggregated["region"] = air_df.groupby(air_df.index // 24)["region"].first().reset_index(drop=True)
df_aggregated["region"] = df_aggregated["region"].apply(lambda x: x.split(" ")[0])
df_aggregated["report_date"] = air_df.groupby(air_df.index // 24)["report_date"].first().astype(str).str[:-2].reset_index(drop=True)

df_aggregated = df_aggregated.groupby(["region", "report_date"])[columns_to_aggregate].mean().reset_index()
df_aggregated = df_aggregated.sort_values(by=['report_date', 'region'], ascending=[True, True]).reset_index(drop=True).reset_index().rename(columns={"index": "id"})
df_aggregated = df_aggregated[new_order]
df_aggregated.info()
```

```python
from sqlalchemy import create_engine, Column, Integer, String, Float, MetaData, Table

# 데이터베이스 연결
engine = create_engine(f'mysql+mysqldb://{DB_USER}:{DB_PASSWORD}@{DB_HOST}/{DB_NAME}')
metadata = MetaData()

# 컬럼 기반으로 동적으로 SQLAlchemy 테이블 생성
columns = []
for col, dtype in zip(df_aggregated.columns, df_aggregated.dtypes):
    # if col == 'id':
    #     columns.append(Column(col, Integer, primary_key=True, autoincrement=True))
    if dtype == 'int64':
        columns.append(Column(col, Integer))
    elif dtype == 'float64':
        columns.append(Column(col, Float))
    elif dtype == 'object':
        columns.append(Column(col, String(10)))

# 테이블 생성
table = Table('air_day', metadata, *columns)
metadata.create_all(engine)

# 데이터프레임 삽입
with engine.connect() as conn:
    df_aggregated.to_sql('air_day', conn, if_exists='append', index=False)
```
