```python
from glob import glob

air_files_dir = "poject_data/air"

excel_files = glob(f"{air_files_dir}/*.xlsx")
excel_files
```

```python
import pandas as pd

# CSV 파일들을 읽어 리스트로 저장
dfs = [pd.read_excel(excel_file) for excel_file in excel_files]

# 모든 DataFrame을 병합하여 하나의 DataFrame 생성
merged_df = pd.concat(dfs, ignore_index=True)

# 병합된 DataFrame 확인
merged_df.info()
```

```python
rename_info = {"지역": "region", "측정일시": "report_date"}
merged_df.rename(columns=rename_info, inplace=True)
merged_df
```

```python
air_df = merged_df.sort_values(by=['report_date', 'region'], ascending=[True, True]).reset_index(drop=True)
air_df.head()
```

```python
drop_col_list = ['망', '측정소코드', "측정소명", "주소"]
air_df.drop(columns=drop_col_list, inplace=True)
```

```python
air_df["region"] = air_df["region"].apply(lambda x: x.split(" ")[0])
air_df.head()
```

```python
new_order = ["id", "region", "report_date", "SO2", "CO", "O3", "NO2", "PM10", "PM25"]

# SO2, CO, O3, NO2, PM10, PM25만 평균값 계산
columns_to_aggregate = ["SO2", "CO", "O3", "NO2", "PM10", "PM25"]
df_aggregated = air_df[columns_to_aggregate].groupby(air_df.index // 24).mean().reset_index(drop=True)

df_aggregated["region"] = air_df.groupby(air_df.index // 24)["region"].first().reset_index(drop=True)
df_aggregated["region"] = df_aggregated["region"].apply(lambda x: x.split(" ")[0])
df_aggregated["report_date"] = air_df.groupby(air_df.index // 24)["report_date"].first().astype(str).str[:-2].reset_index(drop=True)

df_aggregated = df_aggregated.groupby(["region", "report_date"])[columns_to_aggregate].mean().reset_index()
df_aggregated = df_aggregated.sort_values(by=['report_date', 'region'], ascending=[True, True]).reset_index(drop=True).reset_index().rename(columns={"index": "id"})
df_aggregated = df_aggregated[new_order]
df_aggregated.info()
```
