```python
import pandas as pd

# Excel 파일 경로
file_path = "국민관심질병_12개/국민관심질병_요양기관소재지별현황_202303_202405.xlsx"

# workbook = load_workbook(filename=file_path)
# sheet = workbook.active  # 활성화된 시트 선택
# Excel 파일 읽기
df = pd.read_excel(file_path, sheet_name=None)

# 모든 시트의 데이터를 하나의 DataFrame으로 병합
all_data = pd.DataFrame()
for sheet_name, sheet_data in df.items():
    # 첫 번째 행을 컬럼 이름으로 설정
    sheet_data.columns = sheet_data.iloc[0]
    sheet_data = sheet_data[1:]

    # 날짜 컬럼 추가
    sheet_data['진료년월'] = sheet_name

    # DataFrame 병합
    all_data = pd.concat([all_data, sheet_data], ignore_index=True)

# 결과 확인
all_data.head()
```

```python
import pandas as pd
import openpyxl

# Excel 파일 경로

# Excel 파일 읽기
workbook = openpyxl.load_workbook(file_path)
sheet = workbook.active

# A1 셀에서 날짜 정보 추출
date_list = sheet['A1'].value.split('_')[-2:]
start_date = pd.to_datetime(date_list[0], format='%Y%m')
end_date = pd.to_datetime(date_list[1], format='%Y%m')
date_range = pd.date_range(start=start_date, end=end_date, freq='MS').strftime('%Y%m').tolist()
date_range

# 데이터프레임 읽기 (헤더는 3번째 행부터 시작)
df = pd.read_excel(file_path, header=3)

# 빈 DataFrame 생성
columns = [
    "항목", "요양기관소재지구분", "환자수", "내원일수", "청구건수", 
    "요양급여비용총액", "보험자부담금", "진료년월"
]
all_data = pd.DataFrame(columns=columns)
default_num = 2 + 5
default_df = df.iloc[:, 0:default_num]
all_data = pd.concat([all_data, default_df], ignore_index=True, axis=0)
all_data['진료년월'] = date_range[0]
date_range.pop(0)  # 0번째는 default_df에 이미 추가했으므로 제거

# 결과 확인
all_data
```

```python
# 각 달마다 5개의 컬럼이 있음 (환자수, 내원일수, 청구건수, 요양급여비용총액, 보험자부담금)
num_columns_per_month = 5
month_columns = [
    "환자수", "내원일수", "청구건수", "요양급여비용총액", "보험자부담금"
]

start_col = default_num + 0 * num_columns_per_month
end_col = start_col + num_columns_per_month
month_df = pd.DataFrame(columns=columns)
month_df = df.iloc[:, start_col:end_col]
month_df.columns = month_columns
month_df['진료년월'] = date_range[0]
month_df['항목'] = default_df['항목']
month_df['요양기관소재지구분'] = default_df['요양기관소재지구분']
all_data = pd.concat([all_data, month_df], ignore_index=True, axis=0)

# 결과 확인
all_data
```
